<h1>Pre-Processing</h1>
<h2>Tokenization</h2>
For our tokenization task, let's use the nltk TweetTokenizer ...

```python
import re
import nltk
from nltk.tokenize import TweetTokenizer
from nltk.corpus import wordnet
```

```python
def remove_utf(text):
    return re.sub(r'[^\x00-\x7f]',r' ',text)

def remove_tinyURL(text):
    return re.sub(r'http\S+',r'',text)

path = "C:\\tmp\\Tweet-3000.txt"

tweets = ""
file_input = open (path,"r")
lines = file_input.readlines()
for line in lines:
    tweets += (remove_utf(line.lower()))
file_input.close()

print (len(tweets))
```

```python
tweet_tokenizer = TweetTokenizer()
tokens_raw = tweet_tokenizer.tokenize(tweets.lower())
```

```python
print (tokens_raw)
```

We also want to replace tweeter elements like user names and url

```python
tokens = []
for token in tokens_raw:
    if (token == " "):
        continue
    elif (token.startswith('http')):
        tokens.append('URL')
    elif (token.startswith('@')):
        tokens.append('USER_NAME')
    else:
        tokens.append(token)

print (tokens)
```

<h2>Lemmatizing</h2>

Once again, try to experiment with the code and try LDA with and without lemmatization to see which case generates better result. I leave this as an exercise.

```python
def get_lemma(token):
    #Return the morphological variant of this word
    lemma = wordnet.morphy(token)

    if lemma is None:
        return token
    else:
        return lemma

lemmas = [get_lemma(token) for token in tokens]

print (lemmas)
   
```

<h2>Stop-Wording</h2>
Even in this case we will use the NLTK stopwords english dictionary
We will also remove tokens with lenght less than 4 characters

```python
stop_words= set(nltk.corpus.stopwords.words('english'))
tokens_clean = [token for token in lemmas if (len(token)>4 and token not in stop_words)]

print (tokens_clean)
```

<h1> Putting everything together ... </h1>
Ok, now that our pre-processing step work, let's create a single function that, giving a post, returns the cleaned version of that post.

In our case we will also remove URLs and USER_NAME

```python
def clean_tweets(text):
    tokens_raw = tweet_tokenizer.tokenize(text)
    tokens = []
    for token in tokens_raw:
        if (token == " " or token.startswith('http') or token.startswith('@')):
            continue
        else:
            tokens.append(token)
    
    lemmas = [get_lemma(token) for token in tokens]
    tokens_clean = [token for token in lemmas if (len(token)>4 and token not in stop_words)]
    return tokens_clean
```

```python
print(clean_tweets("@GC what a great day today in New_York_City look at http://www.weather.com"))
```

```python
dataset = []
file_input = open (path,"r")
lines = file_input.readlines()
for line in lines:
    dataset.append(remove_utf(line.lower()))
file_input.close()

tweets = [clean_tweets(tweet) for tweet in dataset]
```

```python
print (tweets)
```

<h1>Topic Modeling with LDA</h1>
<h2>Creating a dictionary from the tweets</h2>

We will create a bag-of-words representation of the dictionary
You will get a warning which can be ignored ...

```python
from gensim import corpora
dictionary = corpora.Dictionary(tweets)
corpus = [dictionary.doc2bow(tweet) for tweet in tweets]
```

```python
dictionary.token2id
```

```python
import gensim
k=10
iterations = 20
topic_model = gensim.models.ldamodel.LdaModel(corpus, num_topics=k, id2word = dictionary, passes = iterations)
```

```python
topics = topic_model.print_topics(num_words = 4)
for topic in topics:
    print (topic)
```

<h2>What are the topics in a new tweet?</h2>
With the topic model in place we can take a new tweet and see to which topic it most likely belongs. This can be used as a classification system.

```python
new_tweet = clean_tweets("It seems tomorrow is going to rain :-( ... too bad")
new_tweet_bow = dictionary.doc2bow(new_tweet)
print (topic_model.get_document_topics(new_tweet_bow))
```

It is not easy to determine how many topics using tweet data. You can use LDA more easily in a newsgroup setting, or in a set of articles coming from a well-defined number of topics (sport, art, health, politics, etc.)

<h1> Visualizing Topics </h1>
Let's see how to visualize topics generated by the LDA algorithm

First, you need to install pyLDAvis: read https://pyldavis.readthedocs.io/en/latst/readme.html (only works with Python 3.5).
The library uses PCA to display a multidimentional space into two components

```python
import pyLDAvis.gensim
```

```python
lda_vis = pyLDAvis.gensim.prepare(topic_model,corpus,dictionary,sort_topics=False)
pyLDAvis.display(lda_vis)
```

You are seeing some overlapping (two major overlapping and a partial one), which means we might have less than 10 topics: 7 or 8.
